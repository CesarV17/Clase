```python
import numpy as np
import pandas as pd
df = pd.read_excel('datasets/Canada.xlsx',
                  sheet_name = 'Canada by Citizenship',
                   skiprows = range(20),
                   skipfooter = 2,                
                  engine = 'openpyxl')
df
```

```python
df.drop(['AREA','REG','DEV','Type','Coverage'],
       axis=1,
       inplace = True)

df.rename(columns={'OdName':'Contry','AreaName':'Continent','RegName':'Region'},
         inplace=True)

df.head()
```

```python
df.set_index('Contry', inplace = True)
df.head(2)
```

```python
df['Total'] = df.iloc[:, 4:].sum(axis=1)
df.head(2)
```

```python
years = list (map(str, range (1980,2014)))
years
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.style.use('ggplot')
```

```python
df_continents = df.groupby('Continent', axis = 0).sum()
df_continents
```

```python
df_continents['Total'].plot(kind = 'pie',
                           figsize = (5,6),
                           autopct = '%10.1f%%',
                           startangle=90,
                           shadow=True)
plt.show()
```
