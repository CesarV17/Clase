```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()

df = pd.read_csv('datasets/antropometria-dataset.csv')
df.head()
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.hist(x=df.sistol, bins = 8)
plt.title('Sistol')
plt.subplot(1,2,2)
plt.hist(x=df.diastol,bins=15)
plt.title('Diastol')

plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.hist(x=df.peso, bins = 8)
plt.title('Peso')
plt.subplot(1,2,2)
plt.hist(x=df.cintura,bins=8)
plt.title('Cintura')

plt.show()
```

```python
plt.scatter(x=df.sistol,
           y=df.peso)
plt.show()
```
